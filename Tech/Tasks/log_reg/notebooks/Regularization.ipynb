```python
from sklearn.linear_model import LinearRegression, Ridge,Lasso
from sklearn.metrics import mean_squared_error ,mean_absolute_error, r2_score
from sklearn.model_selection import GridSearchCV ,cross_val_score
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X, y = make_regression(n_samples=100, n_features=10, noise=10, random_state=42)
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_ture = np.array([100,200,300,400,350])
y_pred = np.array([99,201,301,399,390])
```

```python
r2_score(y_ture, y_pred)
```

```python
np.sqrt(mean_squared_error(y_ture,y_pred))
```

```python
mean_absolute_error(y_ture,y_pred)
```

# 1: k-fold cross-validation

```python
lr_scores = cross_val_score(lr, X_train, y_train, cv=5, scoring='r2')
```

```python
lr_scores
```

```python
print(f"{np.mean(lr_scores)*100:.2f}")
```

```python
ridge = Ridge()
```

```python
param_grid = {'alpha': list(range(5,30))}
```

# 2: GridSearchCV

```python
grid_search = GridSearchCV(ridge, param_grid, cv=5, scoring='r2')
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_['alpha']
```

```python
grid_search.best_score_
```
