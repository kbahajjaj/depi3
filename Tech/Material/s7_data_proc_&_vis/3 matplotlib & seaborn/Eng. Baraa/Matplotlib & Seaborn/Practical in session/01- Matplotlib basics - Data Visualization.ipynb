## 3 methods of plotting

```python
#1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1])  # (left, bottom, width, height)
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended one) / Object Oriented API
fig, ax = plt.subplots() #create figure and subplots
ax.plot(x, y);
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
plt.show()
```

# Matplotlib example workflow

```python
# 0. import matplolib and get it ready for plotting in Jupyter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

```python
# 2. Set up Plot
fig, ax = plt.subplots(figsize = (5, 5)) #(width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title = "Simple Plot",
           xlabel = "x-axis",
           ylabel = "y-axis")

# 5. Save & Show with others (you save the whole figure)
fig.savefig("sample-plot.png")
fig.show()
plt.show()
```

# 1) Making figures with NumPy arrays
- Line Plot
- Scatter Plot
- Bar Plot
- Histogram
- Subplot

```python
import numpy as np
```

```python
# create data
x = np.linspace(0, 10, 100) #create 100 of evenly distributed data between 0 and 10
x
```

```python
# Set up & Plot the data & create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
plt.show()
```

```python
# create some data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Nut Butter Store",
          ylabel = "Price");
plt.show()
```

```python
#horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Nut Butter Store",
          xlabel = "Price");
plt.show()
```

```python
#Histogram
# make some data
x = np.random.randn(1000) #normal distributed random data

fig, ax = plt.subplots()
ax.hist(x);
plt.show()
```

## Two options for subplots


```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,ncols = 2,figsize=(10,5))
#fig, (ax1, ax2) = plt.subplots(nrows = 1,ncols = 2,figsize=(10,5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
plt.show()
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                                 ncols = 2,
                                  figsize = (10, 5))

#plot to each different indexes
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
plt.show()
```

# 2) Plotting from Pandas Data Frame

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
car_sales.head(2)
```

```python
#remove $ , . from "Price" column
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "")
#remove last 2 zeros
#car_sales["Price"] = car_sales["Price"].str[:-2]
#car_sales
car_sales
```

```python
car_sales.head()
```

```python
car_sales["Sales Date"] = pd.date_range("5/1/2023",   periods = len(car_sales))
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$,\.]', '', regex=True).astype(int)
```

```python
car_sales.head()
```

```python
car_sales.info()
```

```python
#plot the total sales
car_sales.plot(x="Make", y="Price");
plt.show()
```

```python
#scatter plot
car_sales.plot(x="Make", y="Price", kind="scatter")
plt.show()
```

```python
#Bar Graph
x = np.random.rand(10, 4)

#turn into data frame
df = pd.DataFrame(x, columns=["A", "B", "C", "D"])
df
```

```python
df.plot.bar();
plt.show()
```

```python
df.plot(kind="bar");
plt.show()
```

```python
car_sales.head()
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="bar");
plt.show()
```

```python
#average odometer by make
odometer_df = car_sales.groupby("Make")[["Odometer (KM)"]].sum()
odometer_df.head()
```

```python
odometer_df = car_sales.groupby("Make")[["Odometer (KM)"]].count()
odometer_df.head()
```

```python
odometer_df = car_sales.groupby("Make")[["Odometer (KM)"]].mean()
odometer_df.head()
```

```python
odometer_df.index
```

```python
odometer_df.reset_index().plot(x="Make", y="Odometer (KM)", kind="bar")
plt.title("Odometer(KM) by Make")
plt.show()
```

```python
# Histogram
car_sales["Odometer (KM)"].plot.hist();
plt.show()
```

```python
car_sales["Odometer (KM)"].plot(kind = "hist");
plt.show()
```

```python
#bins
car_sales["Odometer (KM)"].plot.hist(bins=20);
plt.show()
```
