```python
df['age'].nunique()
```

```python
# Matplotlib
plt.figure(figsize=(14,6)) # width , height 
plt.hist(df['age'] , bins = 47 , edgecolor = 'black' , color = 'red' , rwidth = 0.9 , align = 'mid') # know max - min = bin 
plt.xticks([i for i in range(18,65)])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn
plt.figure(figsize=(14,6))
sns.histplot(df['age'] , bins = 47 , edgecolor = 'green' , color = 'red', fill = False , kde = True)
plt.xticks([i for i in range(18,65)])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn
plt.figure(figsize=(14,6))
sns.kdeplot(df['age'] , color = 'red', fill = True)
plt.xticks([i for i in range(18,65,2)])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn
plt.figure(figsize=(14,6))
sns.boxplot(data = df , x = 'age')
plt.xticks([i for i in range(18,65,2)])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df['sex'].value_counts()
```

```python
# pandas
df['sex'].value_counts().plot(kind='bar')
```

```python
df['sex'].value_counts().plot(kind='pie')
```

```python
# matplotlib
plt.figure(figsize=(14,6))
plt.bar(x = list(df['sex'].value_counts().index) , height = df['sex'].value_counts() , color = ['red', 'green'] ,width = 0.5)
plt.title('Gender')
plt.xlabel('sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# matplotlib
plt.figure(figsize=(14,6))
plt.pie(x = df['sex'].value_counts() ,labels = df['sex'].value_counts().index ,explode = [0.1,0.1] , autopct = '%1.2f%%' )
plt.legend (loc='upper right')
plt.title('Gender')
plt.xlabel('sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn
plt.figure(figsize=(14,6))
sns.barplot(x = df['sex'].value_counts().index ,y =df['sex'].value_counts())
plt.title('Gender')
plt.xlabel('sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn
plt.figure(figsize=(14,6))
sns.countplot(data = df , x = 'sex')
plt.title('Gender')
plt.xlabel('sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# subplot
plt.figure(figsize=(14,6))
plt.subplot(2,2,1)
sns.countplot(data = df , x = 'sex')
plt.title('Gender')
plt.xlabel('sex')

plt.subplot(2,2,2)
sns.kdeplot(df['bmi'] , fill = True)
plt.title('Bmi')
plt.xlabel('bmi')

plt.subplot(2,2,3)
sns.boxplot(data = df , x = 'age')
plt.title('Age')
plt.xlabel('age')

plt.subplot(2,2,4)
plt.pie(x = df['sex'].value_counts() ,labels = df['sex'].value_counts().index ,explode = [0.1,0.1] , autopct = '%1.2f%%' )
plt.legend (loc='right')
plt.title('Gender')
plt.xlabel('sex')

plt.tight_layout()
plt.show()
```

```python
# seaborn
plt.figure(figsize=(14,6))
sns.countplot(data = df , x = 'sex' , hue = 'smoker')
plt.title('Gender')
plt.xlabel('sex')
```

```python
corr = df.loc[:, ['age' , 'charges']].corr()
corr
```

```python
# seaborn
plt.figure(figsize=(14,6))
sns.scatterplot(data = df , x = 'age' , y = 'charges')
plt.title('age by charges')
plt.xlabel('age')
plt.xlabel('charges')
plt.show()
```

```python
# seaborn
plt.figure(figsize=(14,6))
sns.jointplot(data = df , x = 'age' , y = 'charges' , kind = 'scatter')
plt.title('age by charges')
plt.xlabel('age')
plt.xlabel('charges')
plt.show()
```

```python
# seaborn
plt.figure(figsize=(14,6))
sns.heatmap(corr , annot= True , cmap = 'Blues')
plt.show()
```
