## Matplotlib example workflow
##### 0. import matplolib and get it ready for plotting in Jupyter
##### 1. Prepare data
##### 2. Set up Plot
##### 3. Plot Data
##### 4. Customize Plot
##### 5. Save & Show with others (you save the whole figure)


```python
import matplotlib.pyplot as plt

%matplotlib inline
 #1 
x = [1,2,3,4]
y=[11,22,5,44]
fig ,ax = plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.set(title= "Figure 1",xlabel = "x-axis",ylabel="y-axis" )
fig.savefig("test.png")
```

# 1) Making figures with NumPy arrays
- Line Plot
- Scatter Plot
- Bar Plot
- Histogram
- Subplot

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
x = np.linspace(0,10,100)
fig, ax  =plt.subplots()
ax.plot(x,x)
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.cos(x))
```

```python
computer_price = {"hp": 10,"mac":8,"apple": 12}
fig,ax = plt.subplots()
ax.bar(computer_price.keys(),height=computer_price.values())
ax.set(title = "Computer Price",xlabel= "Compny Name",ylabel="Price")
```

```python
computer_price = {"hp": 10,"mac":8,"apple": 12}
fig,ax = plt.subplots()
ax.barh(list(computer_price.keys()),list(computer_price.values()))
ax.set(title = "Computer Price",ylabel= "Compny Name",xlabel="Price")
```

## Two options for subplots


```python
fig ,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2 , ncols =2 ,figsize = (10,5) )
x = [1,2,3,4]

ax1.plot(x,x);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.barh(list(computer_price.keys()),list(computer_price.values()));
ax4.hist(np.random.random(1000));
```

# 2) Plotting from Pandas Data Frame

```python
import pandas as pd
car_sales= pd.read_csv("car-sales.csv")
car_sales.info()
```

```python
car_sales["Price"]= car_sales["Price"].str.replace("[$.,]","")
```

```python

car_sales["Price"]= car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype("int")
```

```python
car_sales["sales Date"] = pd.date_range("5/1/2023",periods= len(car_sales))
```

```python
car_sales
```

```python
car_sales.plot(x= "Make",y="Price")
```

```python
car_sales.plot(x= "Make",y="Price",kind= "bar")
```

```python
odDf= car_sales.groupby("Make")[["Odometer (KM)","Price"]].count()
odDf.index
```

```python
odDf
```

```python
odDf.reset_index()
```

```python
odDf.reset_index().plot(x="Make",y="Odometer (KM)" ,kind= "bar" )
```

```python
odDf.reset_index().plot(x="Make",y="Price" ,kind= "bar" )
```

# Setting colors, linewidths, linetypes
- #### possible linestype options `--, , -., :, steps`
- #### ax.plot(x, x, color='#e803fc', label='X vs X', linewidth=1.5, linestyle='--') #RGB Hexcode or CSS Colors
- #### plt.legend()

```python
# fig =plt.figure()
# ax = fig.add_axes([0,0,1,1])

fig,ax = plt.subplots()
x = np.linspace(0,10,10)
ax.plot(x,x,color= "green",ls='dashed',linewidth=3,label = "X vs X", marker ="s" )
ax.plot(x,x/10,color= "red",ls='dashed',linewidth=3 ,label = "X vs X/10")
plt.legend()
```
