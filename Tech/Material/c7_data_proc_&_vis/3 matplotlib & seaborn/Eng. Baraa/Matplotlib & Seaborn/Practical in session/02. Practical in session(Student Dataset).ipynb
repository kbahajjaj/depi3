```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"Cleaned_Student_Dataset.csv")
df
```

```python
plt.plot(df['Student'], df['Math'], marker='o', label='Math')
plt.plot(df['Student'], df['Science'], marker='s', label='Science')

plt.title('Student Grades')
plt.xlabel('Student')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Student'], df['Math'], color='skyblue')
plt.title('Math Scores')
plt.xlabel('Student')
plt.ylabel('Score')
plt.show()
```

```python
sns.boxplot(data=df[['Math', 'Science']])
plt.title('Score Distribution')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, y='Math', x='Gender', palette='pastel')
plt.title("Math Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, y='Science', x='Gender', palette='pastel')
plt.title("Science Score by Gender")
plt.show()
```

```python
sns.scatterplot(data=df, x='Math', y='Science', hue='Gender', s=100)
plt.title('Math vs Science Scores by Gender')
plt.xlabel('Math')
plt.ylabel('Science')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='Math', y='Science', hue='Gender', style='Gender', s=100)
plt.title("Math vs Science Scores")
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.grid(True)
plt.show()

# شرح:
# hue: بيغير لون النقاط حسب الجندر.
#style: بيعطي شكل مختلف للنقطة حسب الفئة.
# s: حجم النقطة.
```

```python
# Dashboard-style layout with multiple plots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Student Performance Dashboard", fontsize=16)

# 1. Histogram of Math Scores
axes[0, 0].hist(df['Math'], bins=6, color='skyblue', edgecolor='black')
axes[0, 0].set_title("Math Score Distribution")
axes[0, 0].set_xlabel("Score")
axes[0, 0].set_ylabel("Number of Students")

# 2. Bar plot for Math Scores per Student
axes[0, 1].bar(df['Student'], df['Math'], color='orange')
axes[0, 1].set_title("Math Scores per Student")
axes[0, 1].set_xlabel("Student")
axes[0, 1].set_ylabel("Score")
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Count plot for Gender
sns.countplot(ax=axes[0, 2], data=df, x='Gender', palette='Set2')
axes[0, 2].set_title("Gender Distribution")
axes[0, 2].set_xlabel("Gender")
axes[0, 2].set_ylabel("Count")

# 4. Boxplot of Math Scores by Gender
sns.boxplot(ax=axes[1, 0], data=df, x='Gender', y='Math', palette='pastel')
axes[1, 0].set_title("Math Score by Gender")

# 5. Scatter plot between Math and Science
sns.scatterplot(ax=axes[1, 1], data=df, x='Math', y='Science', hue='Gender', style='Gender', s=100)
axes[1, 1].set_title("Math vs Science Scores")
axes[1, 1].set_xlabel("Math")
axes[1, 1].set_ylabel("Science")

# 6. Heatmap of correlation
corr = df[['Math', 'Science']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, ax=axes[1, 2])
axes[1, 2].set_title("Subject Correlation")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Line Plot
axes[0, 0].plot(df['Student'], df['Math'], label='Math', marker='o')
axes[0, 0].plot(df['Student'], df['Science'], label='Science', marker='s')
axes[0, 0].set_title('Line Plot - Scores')
axes[0, 0].legend()

# Bar Plot
axes[0, 1].bar(df['Student'], df['Math'], color='orange')
axes[0, 1].set_title('Bar Plot - Math')

# Box Plot
sns.boxplot(data=df[['Math', 'Science']], ax=axes[1, 0])
axes[1, 0].set_title('Box Plot')

# Scatter Plot
sns.scatterplot(data=df, x='Math', y='Science', hue='Gender', ax=axes[1, 1])
axes[1, 1].set_title('Scatter - Math vs Science')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='Gender', palette='Set2')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.pairplot(df, hue='Gender')
plt.suptitle('Pairwise Comparison of Scores by Gender', y=1.02)
plt.show()
```

```python
# Jointplot showing the relationship between Math and Science with KDE
sns.jointplot(data=df, x='Math', y='Science', hue='Gender', kind='scatter', height=7)

plt.suptitle("Joint Plot: Math vs Science Scores by Gender", fontsize=14, y=1.02)
plt.show()
```
